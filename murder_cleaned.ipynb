```python
# Import dependencies
import pandas as pd
```

```python
# Import homicide data
murders_df = pd.read_csv("Resources/murders_raw.csv", low_memory=False)
murders_df.head()
```

```python
# Filter to California homicides
murders_df = murders_df[murders_df["State"]=="California"]
murders_df.head()
```

```python
# Rename City column County
murders_df.rename(columns={"City":"County"}, inplace=True)
murders_df.head()
```

```python
# Drop irrelevant columns
murders_df = murders_df.drop(columns=['Agency Code', 'Record Source'])
murders_df.head()
```

```python
# Rename columns to remove spaces
murders_df.rename(columns={"Record ID": "RecordId",
                  "Agency Name":"AgencyName",
                  "Agency Type": "AgencyType",
                  "Crime Type": "CrimeType",
                  "Crime Solved": "CrimeSolved",
                  "Victim Sex": "VictimSex",
                  "Victim Age": "VictimAge",
                  "Victim Race": "VictimRace",
                  "Victim Ethnicity": "VictimEthnicity",
                  "Victim Count": "VictimCount",
                  "Perpetrator Sex": "PerpetratorSex",
                  "Perpetrator Age": "PerpetratorAge",
                  "Perpetrator Race": "PerpetratorRace",
                  "Perpetrator Ethnicity": "PerpetratorEthnicity",         
                  "Perpetrator Count": "PerpetratorCount"},
                  inplace=True)
murders_df.head()
```

```python
# Examine shape of dataset
murders_df.shape
```

```python
# Check for null values
murders_df.isnull().sum()
```

```python
# Check datatypes
murders_df.dtypes
```

```python
# Import county demographics data
county_df = pd.read_csv("Resources/county_raw.csv")
county_df.head()
```

```python
# Filter to California demographics
county_df = county_df[county_df["State"]=="California"]
county_df.head()
```

```python
# Drop irrelevant columns
county_df = county_df.drop(columns=['Professional', 'Service', 'Office', 'Construction', 
                                    'Production', 'Drive', 'Carpool', 'Transit', 'Walk', 'OtherTransp',
                                   'WorkAtHome', 'MeanCommute', 'PrivateWork', 'PublicWork', 'SelfEmployed', 'FamilyWork'])
county_df.head()
```

```python
# Examine shape of dataset
county_df.shape
```

```python
# Check for null values
county_df.isnull().sum()
```

```python
# Check datatypes
county_df.dtypes
```

```python
# Export cleaned datasets to CSV
murders_df.to_csv(r'Resources/murders_clean.csv', index=False)
county_df.to_csv(r'Resources/county_clean.csv', index=False)
```
